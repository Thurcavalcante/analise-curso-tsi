```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Carregando o dataset
df = pd.read_csv('alunos_tsi.csv')
df
```

```python
# 1. Distribuição por Turma (Ano de Matrícula)
turma_counts = df['Turma'].value_counts()
turma_counts
```

```python
# Visualização da distribuição
plt.figure(figsize=(8, 6))
sns.barplot(x=turma_counts.index, y=turma_counts.values)
plt.title('Distribuição de Alunos por Turma')
plt.xlabel('Turma')
plt.ylabel('Número de Alunos')
plt.show()
```

```python
# 2. Distribuição por Cidade e Região
# Contagem de alunos por cidade
cidade_counts = df['Cidade'].value_counts()
cidade_counts
```

```python
# Visualização da distribuição por cidade
plt.figure(figsize=(12, 8))
sns.barplot(x=cidade_counts.values, y=cidade_counts.index)
plt.title('Distribuição de Alunos por Cidade')
plt.xlabel('Número de Alunos')
plt.ylabel('Cidade')
plt.show()
```

```python
# Contagem de alunos por região
regiao_counts = df['A regiao em que mora'].value_counts()
regiao_counts
```

```python
# Visualização da distribuição por região
plt.figure(figsize=(8, 6))
sns.barplot(x=regiao_counts.index, y=regiao_counts.values)
plt.title('Distribuição de Alunos por Região')
plt.xlabel('A regiao em que mora')
plt.ylabel('Número de Alunos')
plt.show()
```

```python
# 3. Situação Atual no Curso
situacao_counts = df['Situação da Matricula'].value_counts()
situacao_counts
```

```python
# Visualização da distribuição da situação no curso
plt.figure(figsize=(10, 6))
sns.barplot(x=situacao_counts.index, y=situacao_counts.values, )
plt.title('Distribuição da Situação Atual no Curso')
plt.xlabel('Semestres Cursando')
plt.ylabel('Número de Alunos')
plt.show()
```
