```python
import matplotlib.pyplot as plt

# Supondo que você tenha um DataFrame chamado df_alunos
# Substitua df_alunos pelo nome real do seu DataFrame

# Filtrando o DataFrame para incluir apenas a turma TSI 2021 e alunos que participaram do projeto de pesquisa
alunos_turma_2021_projetos = df_alunos[(df_alunos['Turma'] == 'TSI 2021') & (df_alunos['Projetos'] == 'Sim, projeto de pesquisa')]

# Contando o número de alunos para cada turma
contagem_alunos_por_turma = alunos_turma_2021_projetos['Turma'].value_counts()

# Criando um gráfico de barras
contagem_alunos_por_turma.plot(kind='bar', color='skyblue')

# Adicionando rótulos e título ao gráfico
plt.xlabel('Turma')
plt.ylabel('Número de Alunos')
plt.title('Número de Alunos da TSI 2021 que Participaram do Projeto de Pesquisa')

# Exibindo o gráfico
plt.show()
```
