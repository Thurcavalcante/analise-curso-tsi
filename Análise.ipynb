### Criando coluna de Idade a partir da coluna de Data_Nascimento

```python
from datetime import datetime

# Convertendo a coluna 'DataNascimento' para o tipo datetime
dados['Data_Nascimento'] = pd.to_datetime(dados['Data_Nascimento'], format='%d/%m/%Y')

# Obtendo a data atual
data_atual = datetime.now()

# Calculando a idade em anos
# Calcula a diferença em dias entre a data atual e a data de nascimento,
# em seguida, divide por 365.25 para considerar anos bissextos e usa np.floor para arredondar para baixo.
dados['Idade'] = np.floor((data_atual - dados['Data_Nascimento']).dt.days / 365.25)

# Exibindo o DataFrame resultante
dados[['Nome', 'Data_Nascimento', 'Idade']]
```

```python
# Função que constroe gráfico em barras
def constroi_grafico_barras(variavel):
    contagem = dados[variavel].value_counts().reset_index()
    fig = px.bar(contagem, x=variavel, y='count', title ='Quantidade de Alunos por ' + variavel, text_auto=True )
    fig.show()
```
